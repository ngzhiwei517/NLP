#  Loading Data

```python
import pandas as pd

# Expanded dataset with more rows & columns
data = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Helen", "Ian", "Jane"],
    "Age": [25, 30, 35, 40, 29, 33, 27, 38, 45, 31],
    "Salary": [50000, 60000, 70000, 80000, 90000, 55000, 62000, 71000, 68000, 64000],
    "Department": ["HR", "Finance", "IT", "IT", "Marketing", "HR", "Finance", "Marketing", "Finance", "IT"],
    "Experience (Years)": [2, 5, 8, 10, 4, 3, 6, 9, 7, 5],
    "Location": ["KL", "JB", "Penang", "KL", "JB", "Penang", "KL", "JB", "Penang", "KL"]
})
print(data)
```

# Handling Missing Data

```python
data.loc[2,'Age']=None #label-based indexing # Simulate missing value
data.loc[1,'Name']=None
print(data)
print(data.isna().sum())
```

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
data['Age']=data['Age'].fillna(0)
data['Name']=data['Name'].fillna("Brian")
print(data)
```

The fillna() method is used to replace all NaN (Not a Number) or missing values in the selected column.

# Fill missing values/Drop missing values


```python
# Fill all missing values
data_filled = data.fillna("Unknown")
print(data_filled)
print()

# Or drop rows with missing data
df_dropped = data.dropna()
print(df_dropped)
```

# Row and column manipulations

**Removing rows**

```python
print(data)
#removing 4th row
print()
print('After removing 4th row:')
data.drop(4,inplace=True) #when use inplace means it changes the ori DF directly
print(data)
```

# Rename the rows

```python
data.rename({0:"First",1:"Second"},inplace=True)
print(data)
```

# Adding new column

```python
data["new column"]=[1,2,3,4,5,6,7,8,9]
print(data)
```

# Sorting and Merging

```python
sorted_data = data.sort_values(by=['Salary'])
print(sorted_data)
```

# Merging with new df

```python
#first data frame
print(data)

#Second data frame for merging
bonus_data = pd.DataFrame({
    "Name": ["Alice", "Brian", "Charlie", "David", "Grace", "Helen","Ian"],
    "Bonus": [5000, 6000, 7000, 8000, 6500, 4000,2000]
})

print()
print("Second data frame")
print(bonus_data)

```

```python
print("Merge data frame")
new_df=pd.merge(data,bonus_data)
print(new_df)
```

```python
def free_trip(experience):
  if int(experience)>5:
    return 'yes'
  else:
    return 'no'
new_df["Free trip"]=new_df['Experience (Years)'].apply(free_trip)
print(new_df)
print()
new_df.drop("Free trip",axis=1,inplace=True)
print(new_df)
```
