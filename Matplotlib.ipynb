# Bar chart

```python
import matplotlib.pyplot as plt

students = ['Alice', 'Bob', 'Charlie']
scores = [85, 92, 78]

plt.bar(students, scores)
plt.xlabel('Students')
plt.ylabel('Scores')
plt.title('Test Scores')
plt.show()
```

# Line Chart

üìä What is pyplot.plot()?

The plot() function in Matplotlib's Pyplot module is used to draw line graphs, which show the relationship between two sets of values ‚Äî usually one for the X-axis and one for the Y-axis.

```python
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Line plot of x and y")
plt.show()
```

# Matplotlib Markers

The markers module in Matplotlib helps highlight individual data points on plots, improving readability

Consider the following simple example where we plot a line graph with circular markers:

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]

plt.plot(x, y1, marker='o', linestyle='-', color='r',label='Prime Numbers')
plt.plot(x, y2, marker='D', linestyle='--',color='g', label='Even Numbers')

plt.title('Sample Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()  # Displays the labels
plt.show()
```

# Pie Charts

Use plt.pie() instead of plt.plot().

```python
import matplotlib.pyplot as plt
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 30, 20, 10]
plt.pie(sizes,labels=labels, autopct='%1.2f%%')
#autopct stands for "automatic percentage" ‚Äî it tells Matplotlib how to display the percentage labels on each slice of the pie char
#% ‚Üí This starts a format specifier
#1.1f ‚Üí Format the number as a floating point with 1 decimal
#%% ‚Üí To show a literal percent sign, we type %%
plt.show()
```

# Scatter Plots

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [5, 7, 9, 11, 13]
plt.scatter(x,y)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.show()
```

# Histogram

numpy is a library for numerical operations, especially for handling arrays and random data generation.

```python
import matplotlib.pyplot as plt
import numpy as np
data = np.random.randn(1000)
#This line generates 1,000 random numbers from a standard normal distribution
plt.hist(data, bins=30)
#bins=30 means the data range is divided into 30 intervals (bins),
#showing how many data points fall into each interval.
plt.show()
```
