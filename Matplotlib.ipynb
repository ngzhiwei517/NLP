# Adding value labels on a Matplotlib Bar Chart

```python
#we can use plt.text to display value on the top of the bar
#plt.text(x, y, s, ha)

#s: The string (value label) to be displayed.
#ha: Horizontal alignment (‚Äòcenter‚Äô, ‚Äòleft‚Äô, or ‚Äòright‚Äô) for the label.

import matplotlib.pyplot as plt
# Data for the bar chart
x = ["Engineering", "BSc", "MBA", "Bcom", "BBA", "MSc"]
y = [9330, 4050, 3030, 5500, 8040, 4560]

# Creating bar chart
plt.bar(x, y)

# Adding title and labels
plt.title("College Admission")
plt.xlabel("Courses")
plt.ylabel("Number of Admissions")



# Function to add value labels on top of bars
def add_labels(x, y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i],ha='center')  # Placing text slightly above the bar

#The i corresponds to the position (index) of each bar

add_labels(x, y)
```

# Displaying Value Labels Inside a Rectangular Box

```python
import matplotlib.pyplot as plt

# Function to add labels inside a box
def add_labels(x, y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha='center',
                 bbox=dict(facecolor='orange', alpha=0.5))  # Adding a red box around text

# Matplotlib expects the bbox parameter to be a dictionary that describes the properties of the text box.
# alpha	0.8	Sets transparency (1.0 = opaque, 0 = invisible)

x = ["Engineering", "BSc", "MBA", "Bcom", "BBA", "MSc"]
y = [9330, 4050, 3030, 5500, 8040, 4560]

# Setting figure size
plt.figure(figsize=(10, 5))

# Creating bar chart
plt.bar(x, y)

# Adding value labels
add_labels(x, y)

# Adding title and labels
plt.title("College Admission")
plt.xlabel("Courses")
plt.ylabel("Number of Admissions")

plt.show()
```

# Placing Labels at the Center of Each Bar

```python
#i	        X-coordinate of the text ‚Äî usually the center of the bar
#y[i] // 2	Y-coordinate of the text ‚Äî this places it halfway up the bar

# Function to place labels at the center of each bar
def add_labels(x, y):
    for i in range(len(x)):
        plt.text(i, y[i] // 2, y[i], ha='center')  # Placing text at half the bar height


import matplotlib.pyplot as plt
x = ["Engineering", "BSc", "MBA", "Bcom", "BBA", "MSc"]
y = [9330, 4050, 3030, 5500, 8040, 4560]

# Setting figure size
plt.figure(figsize=(10, 5))

# Creating bar chart
plt.bar(x, y)

# Adding value labels
add_labels(x, y)

# Adding title and labels
plt.title("College Admission")
plt.xlabel("Courses")
plt.ylabel("Number of Admissions")

plt.show()
```

# Matplotlib Markers

The markers module in Matplotlib helps highlight individual data points on plots, improving readability

Consider the following simple example where we plot a line graph with circular markers:

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]

plt.plot(x, y1, marker='o', linestyle='-', color='r',label='Prime Numbers')
plt.plot(x, y2, marker='D', linestyle='--',color='g', label='Even Numbers')

plt.title('Sample Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()  # Displays the labels
plt.show()
```

# Visualizing Multiple Marker Shapes

This code generates a plot showcasing different Matplotlib markers. It iterates through a list of marker styles and displays them on the same x-axis, with each marker positioned along a horizontal line at different y-values.

```python
import matplotlib.pyplot as plt

x = range(1, 11)
markers = ['o', 's', '^', 'v', 'D', '*', '+', 'x']

for i, marker in enumerate(markers):
    plt.plot(x, [i*2]*10, marker=marker, linestyle='')

plt.title('Different Matplotlib Markers')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

# Change the Size of Figures using set_figheight() and set_figwidth()

```python
import matplotlib.pyplot as plt

# values on x-axis
x = [1, 2, 3, 4, 5]
# values on y-axis
y = [1, 2, 3, 4, 5]

# naming the x and y axis
plt.xlabel('x - axis')
plt.ylabel('y - axis')

# plotting a line plot with it's default size
print("Plot in it's default size: ")
plt.plot(x, y)
plt.show()

# plotting a line plot after changing it's width and height
f = plt.figure()
f.set_figwidth(9)    # Set width to 9 inches
f.set_figheight(3)   # Set height to 1 inch


print("Plot after re-sizing: ")
plt.plot(x, y)
plt.show()
```

```python
plt.figure(figsize=(9, 3))
plt.plot(x, y)
plt.title("Using figsize")
plt.show()
```

# Changing the default size of the figure by using **rcParams (runtime configuration parameters)**

We can permanently change the default size of a figure as per our needs by setting the figure.figsize. In this example, we are using Matplotlib to create a line plot with the default size and then adjusts the default figure size to [2, 2].

```python
import matplotlib.pyplot as plt

# values on x-axis
x = [1, 2, 3, 4, 5]
# values on y-axis
y = [1, 2, 3, 4, 5]

# naming the x axis
plt.xlabel('x - axis')
# naming the y axis
plt.ylabel('y - axis')

# plotting a line plot with it's default size
plt.plot(x, y)
plt.show()

# changing the rc parameters and plotting a line plot
plt.rcParams['figure.figsize'] = [6, 6]

plt.plot(x, y)
plt.show()

plt.scatter(x, y)
plt.show()
```

# What are subplots in matplotlib?

Subplots in Matplotlib refer to multiple plots arranged within a single figure.


The matplotlib.pyplot.subplots() method provides an easy way to create a figure with multiple plots.


fig, ax = plt.subplots(2, 2)




*  plt.subplots(2, 2) creates a figure (fig) with a 2x2 grid of subplots (ax), resulting in 4 axes objects.



*  ax is a 2x2 numpy array where each element is a subplot (axes object), which can be accessed like a matrix.



```python
import matplotlib.pyplot as plt
import numpy as np

# Creating subplots
fig, ax = plt.subplots(2, 2)

# Generating data
x = np.linspace(0, 10, 1000)
#np.linspace(0, 10, 1000) generates 1000 equally spaced points between 0 and 10.

#np.linspace(0, 10, 5)
#array([ 0. ,  2.5,  5. ,  7.5, 10. ])

# Plot functions with different styles
ax[0, 0].plot(x, np.sin(x), 'r-.', label='sin(x)')


ax[0, 1].plot(x, np.cos(x), 'g--', label='cos(x)')
ax[1, 0].plot(x, np.tan(x), 'y-', label='tan(x)')
ax[1, 1].plot(x, np.sinc(x), 'c.-', label='sinc(x)')

# Adding legends and showing the figure
for axes in ax.flat:
    axes.legend()

plt.tight_layout()
plt.show()
```

This code creates a 2√ó2 grid of subplots using Matplotlib and plots different mathematical functions (sin, cos, tan, and sinc) with distinct line styles. It generates x values using NumPy and assigns each function to a specific subplot.

np.linspace(0, 10, 1000) means:

üëâ Cut the interval from 0 to 10 into 1000 evenly spaced points
üëâ Then use those points as the x-values

# Line plots in subplots

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for subplots
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Create subplots with line plots
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 4))

axes[0].plot(x, y1, color='blue', label='sin(x)')
axes[1].plot(x, y2, color='green', label='cos(x)')
axes[2].plot(x, y3, color='orange', label='tan(x)')

# Add titles and legends
axes[0].set_title('Sine Function')
axes[1].set_title('Cosine Function')
axes[2].set_title('Tangent Function')

for ax in axes:
    ax.legend()

# Adjust layout for better spacing
plt.tight_layout()

# Display the figure
plt.show()
```

Explanation: This code creates a single-row, three-column subplot layout using Matplotlib, plotting sin(x), cos(x) and tan(x) with distinct colors.
